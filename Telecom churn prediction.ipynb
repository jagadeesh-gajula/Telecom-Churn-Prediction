```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('telecom_churn_data.csv')
df.shape
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
uniqueone=[]
for i in df.columns:
    if df[i].nunique()==1:
        uniqueone.append(i)
        
# we are droping columns with only one unique item since we want variance to extract 
# information, ZERO variance is not encouraged
df=df.drop(uniqueone,axis=1)

#_________________________________________________________________________________________

row_count=df.shape[0]
all_unique=[] # all unique rows 
for i in df.columns:
    if df[i].nunique() == row_count:
        all_unique.append(i)
print(all_unique)


# we will drop all unique colums tough we have high variance we can't find patterns in the data
df=df.drop(all_unique,axis=1)
```

```python
pd.DataFrame((df.isna().sum()/df.shape[0])*100)
```

```python
dtype_float=[]
dtype_int=[]
dtype_obj=[]

for i in df.columns:
    if df[i].dtypes == 'float64':
            dtype_float.append(i)
            continue
     
    if df[i].dtypes == 'int64':
            dtype_int.append(i)
            continue
            
    if df[i].dtypes == 'object':
            dtype_obj.append(i)
            continue
            
    else:
        print(i,df[i].dtypes)
        
dtype_int
```
