## Telecom churn case study

### Problem Statement
#### In telecom industry. Customers have freedom to change their service provider anytime. on average 10-15% people change service providers. Given the fact that it costs 5-10X more to aquire new customer than to reatain the existing customers.
##### retaining high profitable customers is the number one business goal.
##### Predicting Churn will be very helpful for business
##### We need to separate high value customers, create a good prediction model and factors affecting churn

```python
# lets create label for our data 
label_columns = ['total_ic_mou_9','total_og_mou_9','vol_2g_mb_9','vol_3g_mb_9']

# we will store a mean into label column to map it into label later on.
df['label'] = df[label_columns].mean(axis=1)

# creating a function to map labels into 1 and 0
def label_map(x):
    if x > 0:   # we are considering who spent anything more than zero as non churn or ZERO
        return 0
    if x == 0:  # we are considering who spent zero on as churned and ONE
        return 1
    else:       # see of anyone belong to other category
        print(x)


# applying lamba function to map into numbers.        
df['label'] = df['label'].apply(lambda x:label_map(x))
```

```python
# defining a function to print metrics for every other model 
def clf_metrics(y_test,y_pred):
    conf = metrics.confusion_matrix(y_test,y_pred )
    TP = conf[1,1] 
    TN = conf[0,0] 
    FP = conf[0,1] 
    FN = conf[1,0]
    print('Sensitivity : ',TP / float(TP+FN))
    print('Specificity : ',TN / float(TN+FP))
    print('False postive rate : ',FP/ float(TN+FP))
    print('Positive predictive value : ',TP / float(TP+FP))
    print('Negative predictive value : ',TN / float(TN+ FN))
    
    
  
```

```python
# Scale Data
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
target = pca_df_original.label
pca_df = scale.fit_transform(pca_df_original.drop('label',axis=1))

# Balance classes
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=43)
x_over, y_over = smote.fit_resample(pca_df,target)

# PCA dimension decomposition
from sklearn.decomposition import IncrementalPCA
pca_tree = IncrementalPCA(n_components=25)
x_over = pca_tree.fit_transform(x_over)

# splitting data into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.3)

# Importing keras module
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

# designing model 
model=Sequential()
model.add(Dense(30,input_dim=x_train.shape[1],activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(25,activation="relu"))
model.add(Dense(12,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

# compiling model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train,np.array(y_train),epochs=30,
          validation_data=(x_test,np.array(y_test)),
          batch_size=1024)
acc = model.history.history['accuracy']
val_acc = model.history.history['val_accuracy']

# plotting metrics
plt.plot(range(30),acc,color="blue")
plt.plot(range(30),val_acc,color="red")
plt.show()
model.save('keras_model.h5')
```
